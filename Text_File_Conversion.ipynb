```python
#compare the performance (time taken) of C,C++,java,R and python programs for
#1. Convert 200mb,400mb,600mb,800mb,and 1000mb text files to upper case.

import matplotlib.pyplot as plt

file_sizes = [200, 400, 600, 800, 1000]

time_taken = {
    "C": [12, 20, 34, 45, 55],
    "C++": [15, 25, 30, 50, 60],
    "Java": [18, 30, 40, 60, 70],
    "R": [20, 35, 45, 60, 80],
    "Python": [25, 40, 53, 75, 100]
}

plt.figure(figsize=(10, 6))
for lang, times in time_taken.items():
    plt.plot(file_sizes, times, marker='o', label=lang)

plt.title("Text File Conversion to Uppercase - Time Comparison")
plt.xlabel("File Size (MB)")
plt.ylabel("Time Taken (seconds)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
